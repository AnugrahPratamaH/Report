#Exploratory Data Analysis & Data Preparation

```python
# cek member

df_member = df.groupby(['Nama'])['Jenis Kelamin'].first().reset_index()
```

```python
df_member.head(2)

#membuat pie chart / bar chart untuk mengetahui besaran perbandingan antara jenis kelamin

#hitung jumlah per jenis kelamin
count_gender = df_member['Jenis Kelamin'].value_counts()

#create pie chart

plt.figure(figsize=(10,5))
plt.pie(count_gender, labels=count_gender.index, autopct='%1.1f%%',startangle=90,wedgeprops={'width':0.4},
        labeldistance=1.1,
        pctdistance=0.80)
plt.title('Distribusi Pelari by Jenis Kelamin')
plt.show()
```

```python
#Analisa trend aktivitas dari awal hingga september


#ambil total aktivitas by tanggal
df_trend_activity = df.groupby(['Tanggal'])['Nama'].count().reset_index()

#convert tanggal ke datetime & pecah antara bulan dan tanggal aktivitas
df_trend_activity['Tanggal'] = pd.to_datetime(df_trend_activity['Tanggal'])
df_trend_activity['Bulan'] = df_trend_activity['Tanggal'].dt.month


#df_trend_activity['Bulan'] = df_trend_activity['Tanggal'].dt.month_name()
df_trend_activity['Hari'] = df_trend_activity['Tanggal'].dt.day

#grouping jumlah aktivitas di setiap bulannya

df_trend_activity = df_trend_activity.groupby(['Bulan'])['Nama'].sum().reset_index()
df_trend_activity = df_trend_activity.rename(columns={'Nama':'Jumlah Aktivitas'})

#urutkan sesuai bulan terkecil (ASC)
df_trend_activity = df_trend_activity.sort_values(by='Bulan')

#mapping bulan

# Mapping angka → nama bulan
bulan_mapping = {
    1: "Januari", 2: "Februari", 3: "Maret", 4: "April",
    5: "Mei", 6: "Juni", 7: "Juli", 8: "Agustus",
    9: "September", 10: "Oktober", 11: "November", 12: "Desember"
}

#convert bulan numeric to text (nama bulan)

df_trend_activity['Bulan'] = df_trend_activity['Bulan'].map(bulan_mapping)


#plot dengan line chart


plt.figure(figsize=(10,5))
ax = sns.lineplot(data=df_trend_activity, x='Bulan', y='Jumlah Aktivitas', marker="o")

#tambahkan label di setiap titik
for i, row in df_trend_activity.iterrows():
    ax.text(row['Bulan'], row['Jumlah Aktivitas']+1, row['Jumlah Aktivitas'], ha='center', va='bottom', fontsize=10,color='black')

ax.text(
    0.95, 0.95, "⚠ Data bulan September\nbelum full 1 bulan",
    transform=ax.transAxes,          # posisi relatif terhadap axes (0-1)
    ha="right", va="top",            # rata kanan atas
    fontsize=10, color="red",
    bbox=dict(boxstyle="round,pad=0.3", facecolor="white", edgecolor="red")
)

#plt.xticks(range(0,5))
plt.ylim(0,100)
plt.title('Trend Aktivitas Pelari')
plt.show()
```

```python
#trend activity per gender

df_trend_act_gender = df.groupby(['Jenis Kelamin', 'Tanggal'])['Nama'].count().reset_index()

#convert tanggal ke bulan
df_trend_act_gender['Tanggal'] = pd.to_datetime(df_trend_act_gender['Tanggal'])
df_trend_act_gender['Bulan'] = df_trend_act_gender['Tanggal'].dt.month

#mapping column

df_trend_act_gender = df_trend_act_gender.rename(columns={'Nama':'Jumlah Aktivitas'})
df_trend_act_gender = df_trend_act_gender[['Jenis Kelamin','Jumlah Aktivitas','Bulan']]

#sort values

df_trend_act_gender = df_trend_act_gender.sort_values(by='Bulan')

#grouping jumlah aktivitas

df_trend_act_gender = df_trend_act_gender.groupby(['Jenis Kelamin','Bulan'])['Jumlah Aktivitas'].sum().reset_index()

#mapping bulan

# Mapping angka → nama bulan
bulan_mapping = {
    1: "Januari", 2: "Februari", 3: "Maret", 4: "April",
    5: "Mei", 6: "Juni", 7: "Juli", 8: "Agustus",
    9: "September", 10: "Oktober", 11: "November", 12: "Desember"
}

df_trend_act_gender['Bulan'] = df_trend_act_gender['Bulan'].map(bulan_mapping)


df_trend_act_gender
```

```python
#create average pace per person each month

df_pace_runner = df.groupby(['Nama', 'Tanggal'])['Pace_Lari'].mean().reset_index()

df_pace_runner['Tanggal'] = pd.to_datetime(df_pace_runner['Tanggal'])
df_pace_runner['Bulan'] = df_pace_runner['Tanggal'].dt.month

df_pace_runner = df_pace_runner.groupby(['Bulan', 'Nama'])['Pace_Lari'].mean().reset_index()

df_pace_runner = df_pace_runner.sort_values(by='Bulan')

df_pace_runner = df_pace_runner.rename(columns={'Pace_Lari':'Average Pace'})

#mapping bulan

# Mapping angka → nama bulan
bulan_mapping = {
    1: "Januari", 2: "Februari", 3: "Maret", 4: "April",
    5: "Mei", 6: "Juni", 7: "Juli", 8: "Agustus",
    9: "September", 10: "Oktober", 11: "November", 12: "Desember"
}

df_pace_runner['Bulan'] = df_pace_runner['Bulan'].map(bulan_mapping)

df_pace_runner.head()
```

```python
df_pace_runner_avg = df_pace_runner.groupby(['Nama'])['Average Pace'].mean().reset_index()

df_pace_runner_avg = df_pace_runner_avg.sort_values(by='Average Pace')

df_pace_runner_avg
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
from kneed import KneeLocator

#Mapping gender
#add gender with merge another df

df_gender = df[['Nama','Jenis Kelamin']]

df_gender = df_gender.drop_duplicates()

df_pace_runner_avg = df_pace_runner_avg.merge(df_gender, on='Nama', how='left')

df_pace_runner_avg = df_pace_runner_avg[['Nama','Average Pace','Jenis Kelamin']]



```

```python
#Encode Gender

le = LabelEncoder()
df_pace_runner_avg['Jenis Kelamin Label'] = le.fit_transform(df_pace_runner_avg['Jenis Kelamin'])
```

```python
df_pace_runner_avg
```

```python
#fit elbow

x_data = df_pace_runner_avg [['Average Pace','Jenis Kelamin Label']]

inertia = []

K = range(1,10) #coba k dari 1 sampai 9

for k in K:
  kmeans = KMeans(n_clusters = k, random_state=42, n_init="auto")
  kmeans.fit(x_data)
  inertia.append(kmeans.inertia_)



plt.plot(K, inertia, marker='o')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```

```python
# cari elbow


kn = KneeLocator(K, inertia, curve='convex', direction='decreasing')
kn.plot_knee()
plt.show()
```

```python
#K = 2

#kmeans
kmeans = KMeans(n_clusters=2, random_state=42,n_init=10)

#fit data
kmeans.fit(x_data)

#get cluster
labels = kmeans.labels_ #label cluster tiap row
centers = kmeans.cluster_centers_ #koordinat centroid cluster

#simpan hasil

df_pace_runner_avg['Cluster']= labels

print(df_pace_runner_avg)
print("Centroid cluster: \n", centers)
```

#Evaluate Model

```python
print("Inertia: ", kmeans.inertia_)
```

```python
from sklearn.metrics import silhouette_score

score = silhouette_score(x_data, labels)
print("Silhouette Score: ", score)
```

```python
#Kesimpulan
"""
k2: inertia 7.5 silhouette 0.5 catatan: cluster mulai terbentuk
k3: inertia 5.08 silhouette 0.55 catatan: lebih baik cluster lebih jelas
k5: inertia 1.5 silhouette 0.6 catatan: lebih rapat & silhouette makin tinggi

jadi kesimpulannya jika k ditambahkan maka silhouette makin tinggi dan cluster makin bagus karena makin spesifik akan tetapi balik lagi dengan keperluan lebih baik dibagi berapa cluster
"""
```

```python
#Model 2 / case 2

#Clustering each gender

df_pace_runner_male = df_pace_runner_avg[df_pace_runner_avg['Jenis Kelamin']=='Laki-Laki']
df_pace_runner_female = df_pace_runner_avg[df_pace_runner_avg['Jenis Kelamin']=='Perempuan']

df_pace_runner_male['Status'] = np.where(df_pace_runner_male['Average Pace'] <= 7.30,"Meet","Not Meet")
df_pace_runner_female['Status'] = np.where(df_pace_runner_female['Average Pace'] <= 7.30,"Meet","Not Meet")




#Encode Status

le = LabelEncoder()
df_pace_runner_male['Status Label'] = le.fit_transform(df_pace_runner_male['Status'])
df_pace_runner_female['Status Label'] = le.fit_transform(df_pace_runner_female['Status'])

df_pace_runner_male_x = df_pace_runner_male[['Average Pace','Status Label']]
df_pace_runner_female_x = df_pace_runner_female[['Average Pace','Status Label']]
```

```python
df_pace_runner_female
```

```python
df_pace_runner_male_x
```

```python
#Cari elbow untuk setiap gender

inertia_male = []
K_male = range(1,5) #coba k dari 1 sampai 9

for k in K_male:
  kmeans = KMeans(n_clusters = k, random_state=42, n_init="auto")
  kmeans.fit(df_pace_runner_male_x)
  inertia_male.append(kmeans.inertia_)
#

plt.plot(K_male, inertia_male, marker='o')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Male')
plt.show()
```

```python
#recheck elbow
kn = KneeLocator(K_male, inertia_male, curve='convex', direction='decreasing')
kn.plot_knee()
plt.show()
```

```python
#Cari elbow untuk setiap gender

inertia_female = []
K_female = range(1,9) #coba k dari 1 sampai 9

for k in K_female:
  kmeans = KMeans(n_clusters = k, random_state=42, n_init="auto")
  kmeans.fit(df_pace_runner_female_x)
  inertia_female.append(kmeans.inertia_)
#

plt.plot(K_female, inertia_female, marker='o')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Female')
plt.show()
```

```python
#recheck elbow female

kn = KneeLocator(K_female, inertia_female, curve='convex', direction='decreasing')
kn.plot_knee()
plt.show()
#
```

```python
#Modelling KMeans
"""
KMeans Laki-Laki = 2
KMeans Perempuan = 3
"""

K_male = 2
K_female = 3

#kmeans
kmeans_male = KMeans(n_clusters=2, random_state=42,n_init=10)
kmeans_female = KMeans(n_clusters=3,random_state=42,n_init=10)

#fit data
kmeans_female.fit(df_pace_runner_female_x)
kmeans_male.fit(df_pace_runner_male_x)

#get cluster
labels_male = kmeans_male.labels_ #label cluster tiap row
centers_male = kmeans_male.cluster_centers_ #koordinat centroid cluster

labels_female = kmeans_female.labels_ #label cluster tiap row
centers_female = kmeans_female.cluster_centers_ #koordinat centroid cluster
labels_male = kmeans_male.labels_ #label cluster tiap row
centers_male = kmeans_male.cluster_centers_ #koordinat centroid cluster

df_pace_runner_male['Cluster']= labels_male
df_pace_runner_female['Cluster']= labels_female

print("Centroid cluster Pria: \n", centers_male)
print("Centroid cluster Wanita: \n", centers_female)
```
